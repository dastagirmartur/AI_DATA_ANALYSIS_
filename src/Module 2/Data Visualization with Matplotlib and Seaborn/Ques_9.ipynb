```python
import matplotlib.pyplot as plt
import numpy as np
products = ['A', 'B', 'C', 'D']
sales = [250, 150, 300, 450]
heights = [160, 170, 165, 180, 175]
weights = [55, 65, 60, 80, 76]
fig, axes = plt.subplots(1, 2, figsize=(12, 6))
axes[0].bar(products, sales, color='skyblue')
axes[0].set_title('Sales Comparison by Product')
axes[0].set_xlabel('Products')
axes[0].set_ylabel('Sales')
axes[1].scatter(heights, weights, color='green')
axes[1].set_title('Height vs Weight')
axes[1].set_xlabel('Height (cm)')
axes[1].set_ylabel('Weight (kg)')
plt.tight_layout()
plt.show()
```
