```python
import matplotlib.pyplot as plt
import numpy as np
products = ['A', 'B', 'C', 'D']
sales = [250, 150, 300, 450]
plt.bar(products, sales, color='skyblue')
plt.grid(True)
plt.xlabel('Products', fontsize=12, fontfamily='serif')
plt.ylabel('Sales', fontsize=12, fontfamily='serif')
plt.title('Sales Comparison by Product', fontsize=14, fontfamily='serif')
plt.show()
```
