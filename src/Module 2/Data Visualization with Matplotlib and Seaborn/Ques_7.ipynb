```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
data = {'Height': [160, 170, 165, 180, 175], 'Weight': [55, 65, 60, 80, 76], 'Size': [10, 15, 13, 18, 12]}
df = pd.DataFrame(data)
sns.scatterplot(x='Height', y='Weight', size='Size', sizes=(50, 200), data=df)
plt.title('Height vs Weight Scatter Plot with Point Sizes')
plt.show()
```
