```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df = sns.load_dataset('iris')
print(df.head())
print(df.describe())
num_array = df['sepal_length'].to_numpy()
mean_value = np.mean(num_array)
sum_value = np.sum(num_array)
print(f"Mean: {mean_value}, Sum: {sum_value}")
num_array = np.array([1, 2, 3, 4, 5])
variance = np.var(num_array)
std_deviation = np.std(num_array)
print(f"Variance: {variance}, Standard Deviation: {std_deviation}")
filtered_data = num_array[num_array > 3]
print(f"Filtered Data: {filtered_data}")
df_filled = df.fillna(df.median(numeric_only=True))
df_dropped = df.dropna()
df['new_column'] = df['sepal_length'] + df['sepal_width']
grouped_df = df.groupby('species').agg({'sepal_length': 'mean'})
print(grouped_df)
plt.plot(df['sepal_length'])
plt.title('Basic Plot')
plt.xlabel('Index')
plt.ylabel('Sepal Length')
plt.show()
sns.histplot(df['sepal_length'], kde=True)
plt.title('Histogram of Sepal Length')
plt.xlabel('Sepal Length')
plt.show()
sns.boxplot(x=df['sepal_length'])
plt.title('Box Plot of Sepal Length')
plt.xlabel('Sepal Length')
plt.show()
```
