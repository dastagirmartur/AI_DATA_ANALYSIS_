```python
import seaborn as sns
import matplotlib.pyplot as plt
iris = sns.load_dataset('iris')
grouped = iris.groupby('species').mean().reset_index()
melted = grouped.melt(id_vars='species', var_name='feature', value_name='mean')
sns.barplot(x='feature', y='mean', hue='species', data=melted)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
